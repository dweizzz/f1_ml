```python
import pandas as pd
import pickle
import numpy as np
```

```python
df = pd.read_csv('FULL_F1_DF_SMAFEATURES.csv')
```

```python
#for i in df.columns: print(i)
```

```python
df2 = df.drop(columns=['Unnamed: 0','Unnamed: 0.1','unique_key','timestamp'])
```

```python
df2
```

```python
sma_const_nan_mn = df2.loc[df2['podium_SMA_constructor'].isna()]['podium'].mean()
```

```python
sma_nat_nan_mn = df2.loc[df2['podium_SMA_nationality'].isna()]['podium'].mean()
```

```python
df2['podium_SMA_constructor'].fillna(sma_const_nan_mn, inplace=True)
df2['podium_SMA_nationality'].fillna(sma_nat_nan_mn, inplace=True)
```

```python
#df2.to_csv("HOLYv1.csv")
```

```python
holy_template = df2.loc[(df2['season']==2023) & (df2['round']==2)]
holy_template
```

```python
holy_template['round'] = np.zeros(holy_template.shape[0])
holy_template['grid'] = np.zeros(holy_template.shape[0])
holy_template['podium'] = np.zeros(holy_template.shape[0])

for c in holy_template.columns:
    if 'weather' in c:
        holy_template[c] = [False] * holy_template.shape[0]
    elif 'circuit_id' in c:
        holy_template[c] = np.zeros(holy_template.shape[0])
    elif 'SMA' in c:
        holy_template[c] = np.zeros(holy_template.shape[0])
    elif 'avg_placement' in c:
        holy_template[c] = np.zeros(holy_template.shape[0])
```

```python
holy_template.to_csv('2023_PRED_TEMPLATE.csv')
```
