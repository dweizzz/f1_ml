```python
train_df = df[df["season"] < 2014]
val_df = df[(df['season'] > 2014) & (df['season'] <= 2018)]
test_df = df[df['season'] > 2018]
```

### Data preprocessing - change podium to only be winner

```python
def process_df(df):
    y = (df['podium'] == 1)

    return df.drop(columns=['podium', 'driver', 'qualifying_time']), y
```

```python
train_X, train_y = process_df(train_df)

train_X
```

```python
val_X, val_y = process_df(val_df)
val_X
```

```python
clf = rf()

clf.fit(train_X, train_y)
```

```python
preds = clf.predict_proba(val_X)

# print(sum(preds)/len(preds))
val_preds = val_df.copy(deep=True)
val_preds['preds']  = preds[:, 1]
```

```python
per_round = val_preds[val_preds['season'] == 2018].sort_values(by='preds', ascending=False).groupby(['round', 'season']).first()
```

```python
sum(per_round['podium'] == 1)/len(per_round)
```

```python
test_X, test_y = process_df(test_df)
test_X
```

```python
preds = clf.predict_proba(test_X)

# print(sum(preds)/len(preds))
test_preds = test_df.copy(deep=True)
test_preds['preds']  = preds[:, 1]
```

```python
per_round = test_preds[test_preds['season'] == 2019].sort_values(by='preds', ascending=False).groupby(['round', 'season']).first()
```
