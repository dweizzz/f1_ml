### Data preprocessing - change podium to only be winner

```python
def process_df(df):
    y = (df['podium'] == 1)

    return df.drop(columns=['podium', 'driver', 'qualifying_time']), y
```

```python
train_df = df[df["season"] < 2014]
val_df = df[(df['season'] > 2014) & (df['season'] <= 2018)]
test_df = df[df['season'] > 2018]

train_X, train_y = process_df(train_df)

train_X
```

```python
test_X, test_y = process_df(test_df)
```

```python
train_y
```

```python
val_X, val_y = process_df(val_df)
val_X
```

```python
'''
with default params, validation set accuracy is 52%
'''
```

```python
#cross validation for n_estimators parameter holding all esle constant
n_estimators_scores = pd.DataFrame(columns = ['n_estimators','val_accuracy'])


for i in [100,200,300,400,550,600,650,700]:
    clf = rf(random_state = 1, n_estimators = i,verbose = 1, )

    clf.fit(train_X, train_y)
    preds = clf.predict_proba(val_X)

    # print(sum(preds)/len(preds))
    val_preds = val_df.copy(deep=True)
    val_preds['preds']  = preds[:, 1]
    
    per_round = val_preds.sort_values(by='preds', ascending=False).groupby(['season', 'round']).first()
    
    accuracy = sum(per_round['podium'] == 1)/len(per_round)
    print(accuracy)
    
    
    #get training scores:
    
    n_estimators_scores = n_estimators_scores.append({'n_estimators': i,
                                                      'val_accuracy': accuracy}, 
                                                     ignore_index = True)
    



```

```python
n_estimators_scores
```

```python
##Max depth tuning holding all else default
max_depth_scores = pd.DataFrame(columns = ['max_depth', 'md_accuracy'])
for i in [5,8,9,10,11,12,13,15,20,25,30]:
    clf = rf(random_state = 1, max_depth = i,verbose = 1, )

    clf.fit(train_X, train_y)
    preds = clf.predict_proba(val_X)

    # print(sum(preds)/len(preds))
    val_preds = val_df.copy(deep=True)
    val_preds['preds']  = preds[:, 1]
    
    per_round = val_preds.sort_values(by='preds', ascending=False).groupby(['season', 'round']).first()
    
    accuracy = sum(per_round['podium'] == 1)/len(per_round)
    print(accuracy)
    max_depth_scores = max_depth_scores.append({'max_depth': i, 'md_accuracy': accuracy}, ignore_index = True)
    


```

```python
max_depth_scores
```

```python
##Min_sample_split tuning holding all else default
min_sample_split_scores = pd.DataFrame(columns = ['min_sample_split', 'mss_accuracy'])
for i in [2,5,10,15,20,25]:
    clf = rf(random_state = 1, min_samples_split = i,verbose = 1, )

    clf.fit(train_X, train_y)
    preds = clf.predict_proba(val_X)

    # print(sum(preds)/len(preds))
    val_preds = val_df.copy(deep=True)
    val_preds['preds']  = preds[:, 1]
    
    per_round = val_preds.sort_values(by='preds', ascending=False).groupby(['season', 'round']).first()
    
    accuracy = sum(per_round['podium'] == 1)/len(per_round)
    print(accuracy)
    min_sample_split_scores = min_sample_split_scores.append({'min_sample_split': i, 'mss_accuracy': accuracy}, ignore_index = True)
    

```

```python
min_sample_split_scores
```

```python
##min_sample_leafs holding all else constant
min_sample_leafs_scores = pd.DataFrame(columns = ['min_sample_leafs', 'msl_accuracy'])
for i in [1,2,3,4,5,7,10,15,20]:
    clf = rf(random_state = 1, min_samples_leaf = i,verbose = 1, )

    clf.fit(train_X, train_y)
    preds = clf.predict_proba(val_X)

    # print(sum(preds)/len(preds))
    val_preds = val_df.copy(deep=True)
    val_preds['preds']  = preds[:, 1]
    
    per_round = val_preds.sort_values(by='preds', ascending=False).groupby(['season', 'round']).first()
    
    accuracy = sum(per_round['podium'] == 1)/len(per_round)
    print(accuracy)
    min_sample_leafs_scores = min_sample_leafs_scores.append({'min_sample_leafs': i, 'mss_accuracy': accuracy}, ignore_index = True)
    

```

```python
min_sample_leafs_scores
```

```python
#exhaustive matrix search:
n_estimators = [100,650,700]
max_depth = [3,4,8,10]
min_sample_split = [5,10,15,25]
min_sample_leafs = [2,7,10]
combo_mat = np.array(np.meshgrid(n_estimators, max_depth, 
                                 min_sample_split, min_sample_leafs)).T.reshape(-1,4)

combo_scores = pd.DataFrame(columns = ['params', 'accuracy'])



for i in range(0,len(combo_mat)):
    clf = rf(random_state = 1, n_estimators = combo_mat[i][0], max_depth= combo_mat[i][1]
             ,min_samples_split = combo_mat[i][2], min_samples_leaf = combo_mat[i][3],verbose = 1, )
    print(combo_mat[i])
    clf.fit(train_X, train_y)
    preds = clf.predict_proba(val_X)

    # print(sum(preds)/len(preds))
    val_preds = val_df.copy(deep=True)
    val_preds['preds']  = preds[:, 1]
    
    per_round = val_preds.sort_values(by='preds', ascending=False).groupby(['season', 'round']).first()
    
    accuracy = sum(per_round['podium'] == 1)/len(per_round)
    print(accuracy)
    combo_scores = combo_scores.append({'params': combo_mat[i], 'accuracy': accuracy}, ignore_index = True)
    



```
