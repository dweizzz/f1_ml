```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix
import sys
```

```python
df = pd.read_csv('for_logistic.csv')
```

```python
df = df.dropna()
```

```python
train, test = train_test_split(df, test_size=0.2)
x_train = train.drop(['podium','season','round','driver','qualifying_time'], axis = 1)
y_train = train['podium']

x_test = test.drop(['podium','season','round','driver','qualifying_time'], axis = 1)
y_test = test['podium']
```

```python
cls = LogisticRegression(max_iter=10000)
cls.fit(x_train.values, y_train)
```

```python
y_pred = cls.predict(x_test)
# print(type(y_pred))

np.where(y_pred != 0)[0]
```

```python
print(metrics.accuracy_score(y_test, y_pred))
```
